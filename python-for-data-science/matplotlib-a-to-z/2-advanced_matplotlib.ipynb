### Annotations - describe/label the points

```python
### this is a simple example for annotation
### syntax: plt.text(x,y,'name')
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4',fontdict={'size':12,'color':'brown'}) # to change the text size and color
```

```python
batter = pd.read_csv(r'E:\Learn_Data_Science\Python_For_Data_Science\Matplotlib_Folder\Datasets_For_Matplotlib\batter.csv')
batter.head()
```

```python
sample_df = batter.head(100).sample(25,random_state=5) # random_state=5, fixed seed for random numbers that will not change later
sample_df
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
```

- here each dot represent represents each player, but we don't know which dot represents which player
- to solve this problem, we use annotation

```python
plt.figure(figsize=(15,7))

plt.scatter(sample_df['avg'], sample_df['strike_rate'])


for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
### to change the dot size depending on the runs scored

plt.figure(figsize=(15,7))

plt.scatter(sample_df['avg'], sample_df['strike_rate'], s= sample_df['runs']) # s is used for that purpose


for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

### Drawing horizontal and vertical lines on graph

```python
plt.figure(figsize=(15,7))

plt.scatter(sample_df['avg'], sample_df['strike_rate'], s= sample_df['runs']) 

plt.axhline(130, color='red') # for horizontal line
plt.axvline(30, color= 'green')


for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])

plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

### subplots()

```python
### Older method

plt.figure(figsize=(15,7))

plt.scatter(batter['avg'], batter['strike_rate'])

plt.title('Avg_vs_Strike-Rate')
plt.xlabel('Avg')
plt.ylabel('Strike-rate')

plt.show()
```

```python
### Another method - we use this method to draw multiple graph on same figure

fig, ax = plt.subplots(figsize=(15,7))

ax.scatter(batter['avg'], batter['strike_rate'])

ax.set_title('Avg_vs_Stike-rate')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike-rate')

plt.show()
```

```python
### draw avg vs strike rate  and avg vs runs (two graphs) in same figure

plt.subplots(nrows=2, ncols=1)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1,figsize= (10,6), sharex= True) # sharex/y= True, when both axes are same

ax[0].scatter(batter['avg'], batter['strike_rate'], color= 'red')
ax[0].set_title('avg_vs_strike-rate')
ax[0].set_ylabel('Strike Rate')

ax[1].scatter(batter['avg'], batter['runs'], color= 'green')
ax[1].set_title('avg_vs_runs')
ax[1].set_ylabel('Runs')

ax[1].set_xlabel('Avg')
```

```python
batter.sample(3)
```

```python
### plotting four graphs in same figure
fig, ax = plt.subplots(nrows=2, ncols=2,figsize= (10,10)) 

ax[0,0].scatter(batter['avg'], batter['strike_rate'], color= 'red')
ax[0,0].set_title('avg_vs_strike-rate')
ax[0,0].set_ylabel('Strike Rate')
ax[0,0].set_xlabel('Avg')

ax[1,0].scatter(batter['avg'], batter['runs'], color= 'green')
ax[1,0].set_title('avg_vs_runs')
ax[1,0].set_ylabel('Runs')
ax[1,0].set_xlabel('Avg')


ax[0,1].hist(batter['avg'], color= 'blue')
ax[0,1].set_title('Histogram of Avg')
ax[0,1].set_ylabel('Count')
ax[0,1].set_xlabel('Avg')

ax[1,1].hist(batter['runs'], color= 'cyan')
ax[1,1].set_title('Histogram of Runs')
ax[1,1].set_ylabel('Count')
ax[1,1].set_xlabel('Runs')

```

### 3D surface plot

```python
### create 3d plot of x^2 + y^2

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
yy.shape
xx.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')

ax.plot_surface(xx,yy,z, cmap= 'viridis')
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z, cmap= 'viridis')
fig.colorbar(p)   # to use colorbar
```

### Contour plots
- to convert 3D to 2D graphs

```python
### use previous 3D graph to convert into 2D graph

fig = plt.figure(figsize=(12,8))
ax = plt.subplot() #.

p = ax.contour(xx,yy,z, cmap= 'viridis') #.
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot() #.

p = ax.contourf(xx,yy,z, cmap= 'viridis') # this another one - filled contour
fig.colorbar(p)
```
