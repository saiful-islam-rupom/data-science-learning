```python
### Change the label(legend) location
plt.plot(batsman['index'],batsman['V Kohli'],label="Virat Kohli",color= 'blue', linestyle= 'dashed', marker= 'D') 
plt.plot(batsman['index'],batsman['RG Sharma'], label= 'Rohit Sharma', color= '#FF0000', linestyle='dashdot', marker="o")

# plt.legend(loc='best') # by default
# plt.legend(loc='right') # changed it
plt.legend(loc='center left') # changed it

plt.title('Carrer comperison between Virat Kolhi and Rohit Sharma')
plt.xlabel("Seasons")
plt.ylabel('Runs scored')
```

### Limiting axes

```python
price = [48000,54000,57000,49000,47000,45000, 9999999] # there are a problem with small vs big number
year = [2015,2016,2017,2018,2019,2020, 2021]

plt.plot(year,price)
```

```python
### limiting axes
price = [48000,54000,57000,49000,47000,45000, 9999999]
year = [2015,2016,2017,2018,2019,2020, 2021]

plt.plot(year,price)
plt.ylim(0, 100000) # limit the y axis to see small numbers
```

```python
### limiting axes
price = [48000,54000,57000,49000,47000,45000, 9999999] 
year = [2015,2016,2017,2018,2019,2020, 2021]

plt.plot(year,price)
plt.ylim(0, 100000)
plt.xlim(2015, 2019) # limiting y axis for limited range of years
```

### Multiple bar for single category

```python

rec = pd.read_csv(r'E:\Learn_Data_Science\Python_For_Data_Science\Matplotlib_Folder\Datasets_For_Matplotlib\batsman_season_record.csv')
rec
```

```python
plt.bar(rec['batsman'],rec['2015'])
```

```python

plt.bar(rec['batsman'],rec['2015'])
plt.bar(rec['batsman'],rec['2016'])   # this is not right way to plot Multiple bar for single category
plt.bar(rec['batsman'],rec['2017'])
```

```python
### this is right way

plt.bar(np.arange(rec.shape[0]) - 0.2,rec['2015'],label='2015',width=0.2,color='yellow') # pointing the bars -> np.arange(rec.shape[0]) - 0.2 ----on x axis
plt.bar(np.arange(rec.shape[0]),rec['2016'],label='2016',width=0.2,color='red')
plt.bar(np.arange(rec.shape[0]) + 0.2,rec['2017'],label='2017',width=0.2,color='blue')

### syntax for xticks -> plt.xticks(old_thing, new_thing)
plt.xticks(np.arange(rec.shape[0]), rec['batsman']) # change the numbers into player names


plt.xlabel("Batsman")
plt.ylabel("Runs")
plt.title("Batsmen Performance Over the Years")
plt.legend()

plt.show() # to get rid from extra details in output
```

```python
# alternative
plt.bar(np.arange(5) - 0.3,rec['2015'],width=0.3,color='yellow')
plt.bar(np.arange(5),rec['2016'],width=0.3,color='red')
plt.bar(np.arange(5) + 0.3,rec['2017'],width=0.3,color='blue')

plt.xticks(np.arange(5), rec['batsman'])
```

# Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale
- it shows data anti-clockwise

```python
# simple data
data = [23,45,100,20,49]
subjects = ['englisg','science','maths','physics','chemistry']
plt.pie(data,labels=subjects)
```

```python
### dataset

team = pd.read_csv(r'E:\Learn_Data_Science\Python_For_Data_Science\Matplotlib_Folder\Datasets_For_Matplotlib\gayle-175.csv')
team
```

```python
plt.pie(team['batsman_runs'], labels= team['batsman'])
```

```python
### add percentage

plt.pie(team['batsman_runs'], labels= team['batsman'], autopct='%0.1f%%')
```

```python
### specify colors
plt.pie(team['batsman_runs'], labels= team['batsman'], autopct='%0.1f%%', colors=['blue','red','yellow','grey','cyan','green'])
```

```python
### explode and shadow
plt.pie(team['batsman_runs'], labels= team['batsman'], autopct='%0.1f%%', explode=[0.1,0,0,0.5,0,0], shadow=True)
```

```python
### to see available styles

plt.style.available
```

```python
### to use style, you have to write this code anywhere - but it will change all the graph's styles in entire code

plt.style.use('classic')

plt.pie(team['batsman_runs'], labels= team['batsman'], autopct='%0.1f%%') # there are some differences in styles
```
