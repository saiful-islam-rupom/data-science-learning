### Write a function that can return the track record of 2 teams against each other

```python
ipl.sample(2)
```

```python
def track(x,y):
    mask1 = ipl['Team1'] == x 
    mask11 = ipl['Team1'] == y
    mask2 = ipl['Team2'] == y
    mask22 = ipl['Team2'] == x
    return ipl[(mask1+mask11) & (mask2+mask22)]

track("Rajasthan Royals","Gujarat Titans")
```

```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df

def euclidean_dis(row):
    pt_A = row['1st point']
    pt_B = row['2nd point']
    return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5

ed = points_df.apply(euclidean_dis,axis=1) # on dataframe; axis=1 means row wise

points_df['E_Distance'] = ed
points_df
```
