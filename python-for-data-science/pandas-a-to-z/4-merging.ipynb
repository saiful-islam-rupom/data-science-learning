# Concatenation

### pd.concat() - to merge multiple dataframes "vertically" {0}
### pd.concat([df1,df2], axis= 1) - to merge multiple dataframes "horizontally" {1}

```python
pd.concat([nov,dec])
pd.concat([nov,dec]).shape
regs = pd.concat([nov,dec],ignore_index=True) # don't retain the old indexes
```

```python
mul_in_df = pd.concat([nov,dec],keys=['Nov','Dec']) # it is a multi-index df
mul_in_df
```

### Working with multi-index df

```python
### Extract the data of November
mul_in_df.loc['Nov']

### Extract the data of December 2nd student
mul_in_df.loc[('Dec',1)]
```

```python
### Merge df horizontally
pd.concat([nov,dec],axis= 1)
```

# Join/merge - to join 2 tables/df on the basis of single common column - like SQL
##### inner join
##### left join
##### right join
##### outer join

```python
### inner join - only shows common elements of both table

### Add a (name)column in 'regs' from 'students' df
students
regs
students.merge(regs,how='inner',on='student_id')
```

```python
### left join - shows all element of left table and common elements of both table 
courses
courses.merge(regs,how='left',on='course_id')
```

```python
### right join - opposite to the left join
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})

# students = pd.concat([students,temp_df],ignore_index=True)
students
regs

students.merge(regs,how='right',on='student_id')
```

```python
### outer join - shows all the elements of both table/df
students.merge(regs,how='outer',on='student_id')
```

### Find the students who enrolled in both months

```python
com_stu_id = np.intersect1d(nov['student_id'],dec['student_id'])

students[students['student_id'].isin(com_stu_id)]
```

### Find course that got no enrollment

```python
yy = np.intersect1d(courses['course_id'],regs['course_id'])

courses[~courses['course_id'].isin(yy)]



### another way
yy = np.setdiff1d(courses['course_id'],regs['course_id'])

courses[courses['course_id'].isin(yy)]

```

### Find students who did not enrolled in any course

```python
zz = np.setdiff1d(students['student_id'],regs['student_id'])

students[students['student_id'].isin(zz)]
```

### Find top 3 students with most number of enrollments

```python
aa = regs.merge(students,how='inner',on='student_id')

aa['name'].value_counts()
aa['name'].value_counts().head(3)
```

### Find the top 3 students who spent most amount of money on courses

```python
xy = regs.merge(courses,how='inner',on='course_id').merge(students,how='inner',on='student_id')

xy.groupby('name')['price'].sum().sort_values(ascending=False).head(3)
```

### IPL - Find the top 3 stadium with highest sixes/match ratio

```python

matches
delivery

ab = matches.merge(delivery,how='inner',left_on='id',right_on='match_id')

six = ab[ab['batsman_runs'] == 6]
ns_v = six.groupby('venue')['venue'].value_counts()
nm_v = matches['venue'].value_counts()

ns_v/nm_v
(ns_v/nm_v).sort_values(ascending=False).head(3)
```
